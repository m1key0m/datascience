Data science is the field that uses data to understand problems, find patterns, and make predictions or decisions.

* Data collection – getting data from surveys, apps, sensors, websites, etc.
* Data cleaning – fixing missing, messy, or incorrect data 
* Data analysis – looking for patterns and trends 
* Statistics & math – understanding what the data really means 
* Programming – often using Python or R to work with data 
* Machine learning – teaching computers to learn from data and make predictions
* Communication – explaining results using charts, graphs, and clear explanations

```python
df_fisher = pd.read_csv(fisher_data)
df_fisher

df_fisher['Names']

df_fisher['Hours']

df_fisher.shape

df_fisher.isnull() 

df_fisher.isnull().sum() 
```

```python
avg_hours = df_fisher['Hours'].mean()
avg_hours

ave_catch = df_fisher['Catches'].mean()
ave_catch
```

```python
df_fisher[df_fisher.Hours < avg_hours]
```

```python
good_fisher = pd.Series(df_fisher.Catches >= 11)
good_fisher
```

```python
df_fisher = pd.concat([df_fisher, good_fisher.rename('Good')], axis=1)
df_fisher
```

```python
fig = plt.figure(figsize=(8, 6))
plt.bar(x=df_fisher.Names, height=df_fisher.Catches, color="green")
plt.title("Fisher Catches")
plt.xlabel("Name")
plt.ylabel("Catches")
plt.xticks(rotation=90)
plt.show()
```

```python
catches = df_fisher.Catches

fig = plt.figure(figsize=(11,4))

plt.hist(catches)

plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

fig.show()
```

```python
min_val = df_fisher.Catches.min()
print(min_val)

max_val = df_fisher.Catches.max()
print(max_val)

mean_val = df_fisher.Catches.mean()
print(mean_val)

median_val = df_fisher.Catches.median()
median_val

mode_val = df_fisher.Catches.mode()
mode_val
```

```python
import math

mean = df_fisher.Hours.mean()
variance = 0 

for hours in df_fisher.Hours:
    variance = variance + ((mean - hours)**2)

variance = variance / df_fisher.shape[0]
variance

std = math.sqrt(variance)
std
```
