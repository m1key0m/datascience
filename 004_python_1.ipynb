(4). Leap Year

An extra day is added to the calendar almost every four years as February 29, and the day is called a leap day. It corrects the calendar for the fact that our planet takes approximately 365.25 days to orbit the sun. A leap year contains a leap day.
In the Gregorian calendar, three conditions are used to identify leap years:
   a. The year can be evenly divided by 4, is a leap year, unless:
   b. The year can be evenly divided by 100, it is NOT a leap year, unless:
   c. The year is also evenly divisible by 400. Then it is a leap year.
This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years. Source 
Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.
Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function. 
This means that in the Gregorian calendar, the years 2000 and 2400 are leap years, while 1800, 1900, 2100, 2200, 2300 and 2500 are NOT leap years.
Constraints (1900 <= year <= 10**5)

```python
def is_leap(year):
   leap = False
  
   try:
       if ( year >= 1900 ) and ( year <= 10**5  ):
           if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
               leap = True
       else:
           raise ValueError
  
   except ValueError:
       print("Error: Year must be between 1900 and 10 power of 5")

   return leap

year = int(input())
print(is_leap(year))
```

The lowest grade of 37.2 belongs to Tina. The second lowest grade of 37.21 belongs to both Harry and Berry, so we order their names alphabetically and print each name on a new line.

```python
import sys

n = 5
students = [['Harry', 37.21], ['Berry', 37.21], ['Tina', 37.2], ['Akriti', 41], ['Harsh', 39]]

students.sort(key=lambda x: x[0])

# Find the second lowest grade
scores = []
  
for student in students:
   scores.append(student[1])

unique_scores = set(scores)
sorted_scores = sorted(unique_scores)

second_lowest_score = sorted_scores[1]

#print(second_lowest_score)
#sys.exit()

# Print the names of students with the second lowest grade
for student in students:
   if student[1] == second_lowest_score:
       print(student[0])
```
