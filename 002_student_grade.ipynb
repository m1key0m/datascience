```python
def show_density(data):
    fig = plt.figure(figsize=(10,4))
    data.plot.density()

    plt.title('Data Density')

    plt.show()
```

```python
import math

mean = df_fisher.Hours.mean()
variance = 0 

for hours in df_fisher.Hours:
    variance = variance + ((mean - hours)**2)

variance = variance / df_fisher.shape[0]
variance

std = math.sqrt(variance)
std
```
