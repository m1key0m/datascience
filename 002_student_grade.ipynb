This notebook analyizes pass/fail data for class

```python
names = ['Liam', 'Ava', 'Noah', 'Maya', 'Ethan', 'Chloe', 'Lucas', 'Zara', 'Caleb', 'Oscar', 'Lila', 'Felix', 'Aria', 'Marco', 'Elise', 'Rowan', 'Tessa', 'Julian', 'Ivy', 'Milo', 'Sienna', 'Kai', 'Harper', 'Adrian'] 
```
