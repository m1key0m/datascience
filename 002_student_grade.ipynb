```python
np_hours = np.array(hours)
```

```python
np_catches = np.array(catches)
```

```python
df_fisher = pd.DataFrame({
    "Names": np_names,
    "Hours": np_hours,
    "Catches": np_catches
})
```

```python
df_fisher
```

```python
df_fisher.isnull()
```

```python
df_fisher.isnull().sum() 
```

```python
ave_hours = np.nanmean(hours)
ave_hours
```

```python
ave_catches = np.nanmean(catches)
ave_catches
```

```python
df_fisher[df_fisher.Hours > ave_hours]
```

```python
df_fisher[df_fisher.Hours < ave_hours]
```

```python
df_fisher_rep = df_fisher.fillna(df_fisher.mean(numeric_only=True))
df_fisher_rep
```

```python
df_fisher = df_fisher.dropna(axis=0)
df_fisher
```

```python
df_fisher[df_fisher.Hours > ave_hours].Catches.mean()
df_fisher
```

```python
good_fisher = pd.Series(df_fisher.Catches >= 9)
good_fisher
```

```python
df_fisher = pd.concat([df_fisher, good_fisher.rename('Good')], axis=1)
df_fisher
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.bar(x=df_fisher.Names, height=df_fisher.Catches)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 7.2))
plt.bar(x=df_fisher.Names, height=df_fisher.Catches, color="green")
plt.title("Fisher Catches")
plt.xlabel("Name")
plt.ylabel("Inches")
plt.xticks(rotation=90)
plt.show()
```

```python
catches = df_fisher.Catches

fig = plt.figure(figsize=(11,4))

plt.hist(catches)

plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

fig.show()
```

```python
min_val = df_fisher.Catches.min()
print(min_val)
```

```python
max_val = df_fisher.Catches.max()
print(max_val)
```

```python
mean_val = df_fisher.Catches.mean()
print(mean_val)
```

```python
median_val = df_fisher.Catches.median()
median_val
```

```python
mode_val = df_fisher.Catches.mode()
mode_val
```

```python
catches = df_fisher.Catches

fig = plt.figure(figsize=(11,4))

plt.hist(catches)

plt.axvline(x=min_val, color='yellow')
plt.axvline(x=max_val, color='red')
plt.axvline(x=mean_val, color='blue')
plt.axvline(x=median_val, color='green')
plt.axvline(x=mode_val, color='purple')

plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

fig.show()
```

```python
catches = df_fisher.Catches

fig = plt.figure(figsize=(11,4))

plt.boxplot(catches)

plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

fig.show()
```

```python
def show_density(data):
    fig = plt.figure(figsize=(10,4))
    data.plot.density()

    plt.title('Data Density')

    plt.show()
```

```python
show_density(df_fisher.Catches)
```

```python
show_density(df_fisher.Hours)
```
